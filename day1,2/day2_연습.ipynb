### 범위강조연산

```python
import scipy.stats as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img=cv2.imread('lena.png' , cv2.IMREAD_GRAYSCALE)
print(img.shape)
```

```python
def posterlut(f, t):
    return np.array([255 if x>=f and x<=t else x for x in range(256)]).astype('uint8') #max값 128

bimg = cv2.LUT(img, posterlut(125,150))
plt.imshow(np.hstack([img,bimg]), cmap = 'gray')
```
