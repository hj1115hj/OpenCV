```python

img = cv2.imread("ggg.jpg")
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

result = img
result = img.copy()
print(result.shape)
print(result.shape[0])

# 같은 gray 3장 겹쳐놓은것
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        # 자동형변환  integer - > 32bit
        col = (int(img[r,c,0])+ img[r,c,1]+ img[r,c,2]) /3
        result[r,c,0] = col
        result[r,c,1] = col
        result[r,c,2] = col
 
plt.imshow( cv2.cvtColor(result, cv2.COLOR_BGR2RGB) )

# 이범위 안에 있으며 피부색일 확률이 높다.
# h색상 0 ~20
# s채도 48~255
# i명도 80 255

```

```python

img = cv2.imread("ooo.jpg")
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

result = img
result = img.copy()
print(result.shape)
print(result.shape[0])

# 같은 gray 3장 겹쳐놓은것
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if h[r,c] >=0 and h[r,c] <= 20 and s[r,c] >=48 and s[r,c] <=255 :
            if v[r,c] >= 80 and v[r,c] <=255:
                result[r,c,0] = img[r,c,0]
                result[r,c,1] = img[r,c,1]
                result[r,c,2] = img[r,c,2]
        else :
            result[r,c,0] = 0
            result[r,c,1] = 0
            result[r,c,2] = 0
            
plt.imshow( cv2.cvtColor(result, cv2.COLOR_BGR2RGB) )

# 이범위 안에 있으며 피부색일 확률이 높다.
# h색상 0 ~20
# s채도 48~255
# i(v)명도 80 255
```

```python
img =cv2.imread("ppp.jpg")
mask = np.zeros((img.shape[0],img.shape[1],3), dtype = 'uint8')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h,s,v =cv2.split(hsv)
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if h[r,c] >=59 and h[r,c] <= 59+10 and s[r,c] >=246-10 and s[r,c] <=246+10 :
            if v[r,c] >= 255-5 and v[r,c] <=255+5:
                mask[r,c,:] = 0
            
        else :
            mask[r,c,:] = 255
    

all = np.hstack([img, mask, img&mask])
plt.imshow( cv2.cvtColor(all, cv2.COLOR_BGR2RGB) )
```
