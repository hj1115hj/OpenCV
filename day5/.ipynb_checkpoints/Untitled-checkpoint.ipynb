```python
#
image = cv2.imread("./OpenCV/book.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (3, 3), 0)
# 경계선만 검출하는 함수
edged = cv2.Canny(gray, 10, 250)
imshow("",edged)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
imshow("",kernel)

'''
Canny만 쓴 상태에서는 자잘한 것들도 컨투어로 만들어주기때문에 closed 연산을 해야함
'''
closed = cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)
imshow("",closed)
cnts, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
total = 0

for c in cnts:
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    if len(approx) == 4:
        # -1 : 찾은 컨투어 전부다 그려줌, 4는 컨투어의 굵기
        cv2.drawContours(image, [approx], -1, (0, 255, 0), 4)
        total += 1
print("I found {0} books in that image".format(total))
imshow("Output", image)
```

```python
from collections import Counter

def detect_weiqi(img):  
    txt = 'black'
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, threshold = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

    # 0의개수와 255 의 개수를 cnt 가져온것이 0이아니면 255가 가장 많다는것
    c = Counter(list(threshold.flatten()))
    print(c.most_common())
    if c.most_common()[0][0] != 0:
        txt = 'white'
        # threshold는 어떻게 threshold를 했는지
    return txt, threshold


img = cv2.imread('./OpenCV/stone.png')

img = cv2.medianBlur(img, 5)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 해상도,  원간 최소 거리      edge threshold       circle 중심점 histogram수,
# 1,        20,             param1=100,  param2=30, minRadius=10, maxRadius=50
# param1 : edge threshold low ->  edge가 검출되어 같은 원이 검출됨
# param2 : 중심점 histogram 수가 작으면 많은 원이 검출됨

# 가장자리의 기울기를 계산해서 원의 형태를 예측하는 함수
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20, param1=100, param2=30, minRadius=10, maxRadius=50)

circles = np.uint16(np.around(circles))
print(circles)

font = cv2.FONT_HERSHEY_SIMPLEX
for i in circles[0, :]:
    x, y, r = i
    crop_img = img[y - r:y + r, x - r:x + r]    
    txt, threshold = detect_weiqi(crop_img)
    
    if txt == 'black' :  
        # 실제바둑돌에 반지름의 70%만 검정색으로 안을 칠해주기
        cv2.circle(img, ( x, y),int(r*0.7), (0, 0, 0), -1)
    else :
        cv2.circle(img, (x, y), int(r*0.7), (255, 255, 255), -1)            


imshow("", img)
```

```python
#딥러닝에서 이동연산으로 발생하는 빈곳은 인접한 픽셀로 채운다.
img = cv2.imread('./OpenCV/lena.jpg')
# 변환 행렬, x축으로 10 y축으로 20 이동
height, widht = img.shape[:2]
M = np.float32([
    [1,0,30], #x축
    [0,1,100]]) # y축

#딥러닝에서 이동연산으로 발생하는 빈곳은 인접한 픽셀로 채운다.
c= img[0][0]
print(c)
# borderValue: 빈값으로 어떤 색을 넣을지
dst = cv2.warpAffine(img,M,(widht,height), borderValue =(int(c[0]),int(c[1]),int(c[2])))
imshow("",dst)
```

### 랜덤하게 이미지 이동시키기

```python
import random
print(random.randint(1,10))
```

```python

img = cv2.imread('./OpenCV/lena.jpg')
dx = 50
dy = 100
imgcpy = img.copy()
height, widht = img.shape[:2]
for _ in range(0,10) :
    tx=random.randint(1,dx)
    ty=random.randint(1,dy)
    
    M = np.float32([
        [1,0,tx], #x축
        [0,1,ty]]) # y축
    c = img[0][0] # img[0,0,:]와 같은 의미 rgb가져옴
    
    dst = cv2.warpAffine(img,M,(widht,height), borderValue =(int(c[0]),int(c[1]),int(c[2])))
    imgcpy=np.hstack([imgcpy,dst])
imshow("result",imgcpy)
cv2.imwrite("out1.jpg",imgcpy)
```

```python

img = cv2.imread('./OpenCV/lena.jpg')
imgcpy = img.copy()
dx = 50
r = np.random.randint(dx*2,size = (5,2))-dx  #-50 ~+50 사이 수출력

M = np.float32([
    [1,0,0], #x축
    [0,1,0]]) # y축

c = img[0][0]
for p in r.tolist() :
    M[:,2] =p
    dst = cv2.warpAffine(img,M,(widht,height), borderValue =(int(c[0]),int(c[1]),int(c[2])))
    imgcpy=np.hstack([imgcpy,dst])
    
imshow("result",imgcpy)
cv2.imwrite("out1.jpg",imgcpy)
```

```python
height, width = img.shape[:2]
img_center = (width/2, height/2) 
# 회전양/스케일 과 얼마만큼 이동을해야 센터로 올수있는지이동량 나타내는 각도 
'''
[
  cos30 sin30  센터로 오기 위한 이동량
  -sin30 cos30 ..    
]
이미지를 축소시키지 않기 때무에 이미지가 잘리는 것
'''
'''
1 ->0,5 로 변환시 
0.5*[
  cos30 sin30
  -sin30 cos30
    
]
이미지를 축소시켜서 이미지가 잘리지 않도록하는것
'''
#반 시계방향으로 회전하는 각도
M = cv2.getRotationMatrix2D(img_center, 30, 1.0)
print(M)
#(widht,height) : 회전중심축- [] 오른쪽 사각형축으로 축소시키면 이미지가 사라짐? 
rotated_image = cv2.warpAffine(img,M,(width , height), borderValue = (255,255,255) )
print(rotated_image.shape)

imshow("", rotated_image)
```

```python
frame = cv2.imread("./mart2.jpg")
roi = frame.copy()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

gray_blur = cv2.medianBlur(gray, 9)

imshow("gray->median_blur",np.hstack([gray, gray_blur]))

thr, mask = cv2.threshold(gray, 0, 255,
        cv2.THRESH_BINARY+cv2.THRESH_OTSU)


imshow("binary",mask)
#침식후 팽창
opened = cv2.morphologyEx(mask,cv2.MORPH_OPEN,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    
                          ,iterations =10) 
imshow("opend",opened)

cont_img = opened.copy()
contours, _ = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,
                              cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
```

```python
cnt = contours[0]
peri = cv2.arcLength(cnt,True)
approx = cv2.approxPolyDP(cnt,0.02* peri, True)
len(approx)
frame_copy = frame.copy()
for a in approx:
    p = tuple(a[0])
    print(p)
    cv2.circle(frame_copy, p, 150, (255,0,0), -1)
    imshow('',frame_copy)
```
